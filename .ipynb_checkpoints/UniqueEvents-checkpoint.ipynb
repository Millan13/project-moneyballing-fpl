```python
def match_unique_events(match):
    
    '''
    Takes a 5 digit match code, and returns a list of unique events that happened in that match.
    
    '''
    #Get the comments
    info = get_full_match_info(match)
    comments = info['Events']
    comments = [i.split(". ",1)[1] for i in comments]
    
    #Get the players
    teams = info['Players']
    players = []
    for i in ['HomeStarts','HomeSubs','AwayStarts','AwaySubs']:
        players += teams[i]
    
    #Return the set of unique events, which have been sanitised
    return set([event_sanitiser(i,players = players) for i in comments])



def unique_events(matches):
    '''
    Takes a list of match codes, and returns the set of unique events that happend in those matches.
    
    '''
    #Initiate an empty set
    event_set = set()
    
    #Iterate through matches
    for i in matches:
        new_events = match_unique_events(i)
        
        #Take the union of events we already know, and the events in this match
        event_set = event_set.union(new_events)
        
    return event_set


def assist_split(event):
    
    split = event.split('. Assisted')
    
    if len(split) > 1:
        split[1] = 'Assisted'+split[1]
    
    for i in range(len(split)):

        if split[i][-1] != '.':
            split[i] += '.'
            
    split = [i.replace('PLAYER1','PLAYER'
                      ).replace('PLAYER2','PLAYER'
                               ).replace('TEAM1','TEAM'
                                        ).replace('TEAM2','TEAM') for i in split]
    
    return split
```
